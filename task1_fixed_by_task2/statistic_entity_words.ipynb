```python
import json
train_tagged_file = "./experiment/linked_data/train.fixed.tok.lc.tagged.en.json"
with open(train_tagged_file, 'r') as fp:
    train_tagged_data = json.load(fp)
```

```python
train_part_ids_file = "/home/xhuang/work/corpus/multi30k/original/splits/train_images.txt"
with open(train_part_ids_file, 'r') as fp:
    train_ids = [i.strip().split(".")[0] for i in fp.readlines()]
print(train_ids[0])
```

```python
print(train_tagged_data[train_ids[0]])
```

```python
p_tagged_words = {}
w_tagged_words = {}
for iid in train_ids:
    tagged_sentence = train_tagged_data[iid]
    sent = tagged_sentence['sentence'].strip().split()
    for w, p_tag, w_tag in zip(sent, tagged_sentence['all_tag'], tagged_sentence['key_tag']):
        if p_tag is not None:
            if w in p_tagged_words:
                p_tagged_words[w] += 1
            else:
                p_tagged_words[w] = 1
        if w_tag is not None:
            if w in w_tagged_words:
                w_tagged_words[w] += 1
            else:
                w_tagged_words[w] = 1
```

```python
print(len(w_tagged_words), len(p_tagged_words))
```

```python
print(w_tagged_words)
```

```python
phrase_lengths = {}
for iid in train_ids:
    tagged_sentence = train_tagged_data[iid]
    for phrase_item in tagged_sentence['phrases']:
        phrase = phrase_item['phrase'].strip().split()
        len_p = len(phrase)
        if len_p in phrase_lengths:
            phrase_lengths[len_p] += 1
        else:
            phrase_lengths[len_p] = 1
            
print(phrase_lengths)
```

```python
import json
with open("tagged_key_words.json", 'w', encoding='utf-8') as fp:
    json.dump(w_tagged_words, fp)
with open("tagged_phrase_words.json", 'w', encoding='utf-8') as fp:
    json.dump(p_tagged_words, fp)
```
